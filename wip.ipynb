```python
import datetime
import pandas as pd
from datetime import datetime
import requests
import json
```

```python
item_list = "Resources\items.json"
item_list_df = pd.read_json(item_list, encoding= "ISO-8859-1")
item_list_df

```

```python
# Initialize empty lists to store extracted data
item_ids = []
item_values = []

# Extract and append data
for item_id, item_data in item_list_df.items():
  # Extract item ID (dictionary key)
  item_ids.append(item_id)
  
  # Extract value for "en" key (assuming English)
  if "en" in item_data:  # Check if "en" key exists
    item_values.append(item_data["en"])
  else:
    item_values.append("N/A")  # Handle missing English key (optional)

# Create DataFrame
item_id_df = pd.DataFrame({"ID": item_ids, "English Value": item_values})


# Define the search function using a lambda function
search_id = lambda name: item_id_df[item_id_df["English Value"] == name]["ID"].iloc[0]  # Assuming unique names

# Get user input for name to search
input_name = input("Enter the name to search for (English value): ")

# Try searching for the ID and handle potential errors
try:
  id_found = search_id(input_name)
  print(f"ID for '{input_name}': {id_found}")
except IndexError:
  print(f"Error: Item with name '{input_name}' not found in the DataFrame.")

id_found
```

```python


url = "https://universalis.app/api/v2/"
world = input("What world are you searching in? (ie Gilgamesh, Siren, Cactaur...)")






universalis_url = f"{url}{world}/{id_found}"
response_data = requests.get(universalis_url).json()
print(json.dumps(response_data, indent=4))
```

```python
# Define the desired keys to extract
desired_keys = [input_name, "recentHistoryCount", "averagePriceNQ", "averagePriceHQ", "listingsCount"]

extracted_values = {}
for key in desired_keys:
  if key in response_data:
    extracted_values[key] = response_data[key]
  else:
    extracted_values[key] = None  # Handle missing keys (optional)



# Create response_dataFrame
df = pd.DataFrame([extracted_values])
df
# Print response_dataFrame
```
